```python
import pandas as pd
```

```python
file = "data/wyniki_gl_na_kandydatow_po_gminach_w_drugiej_turze_utf8.csv"
```

```python
df = pd.read_csv(
    file,
    sep=";",
    quotechar='"',
    dtype={
        "TERYT Gminy": str
    },
    na_values=["", " "],
    keep_default_na=True,
    engine="python"
)
```

```python
df.head()
```

```python
num_cols = [c for c in df.columns if c not in ("TERYT Gminy", "Gmina", "Powiat", "Województwo")]
df[num_cols] = df[num_cols].apply(pd.to_numeric, errors="coerce")
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
df.info()
```

```python
df = df[df["TERYT Gminy"].notna()]
df["TERYT Gminy"] = df['TERYT Gminy'].str.zfill(5)
df = df.set_index("TERYT Gminy", verify_integrity=True)
print("Ilość unikalnych gmin:", df["Gmina"].unique())
```

```python
df.to_csv("wybory_czyste.csv", index=True, encoding='utf-8-sig')
```
