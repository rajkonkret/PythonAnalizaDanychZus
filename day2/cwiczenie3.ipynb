```python
store1 = pd.read_csv("data/Store1.csv")
store1.head()
```

```python
store2 = pd.read_csv("data/Store2.csv")
store2.head()
```

```python
sales_store1 = (store1[['Grocery_sales', 'Stationary_sales']].sum()).sum()
sales_store1
```

```python
month_sales1 = store1.groupby(['Months'])[['Grocery_sales', 'Stationary_sales']].agg('sum').agg('sum', axis=1)
month_sales1
```

```python
month_sales1 = store1.groupby(['Months'])[['Grocery_sales', 'Stationary_sales']].agg('sum').agg('sum', axis=0)
month_sales1
```

```python
store_greater = store1[store1['Stationary_sales'].gt(store2['Stationary_sales'])]
store_greater
```
